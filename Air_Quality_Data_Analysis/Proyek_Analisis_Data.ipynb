### Cleaning Data

**Dengan cara yang sama dengan sebelumnya sedikit melakukan modifikasi yaitu pengecekan tipe data nya apa
dilakukan pengisiian nilai null**

```python
for i, df_list_ in enumerate(df_list):
    print(f"Filling NULL values in dataset {df_names[i]}:")
    for column in df_list_.columns:
        if df_list_[column].dtype in ['float64', 'int64'] and column != "RAIN":
            mean_value = df_list_[column].mean()
            df_list_[column].fillna(mean_value, inplace=True)
            print(f"Column '{column}': filled with mean value {mean_value:.2f}")
        elif df_list_[column].dtype == 'object' or column == "RAIN":
            mod_value = df_list_[column].mode()[0]
            df_list_[column].fillna(mod_value, inplace=True)
            print(f"Column '{column}': filled with mode value '{mod_value}'")
    print('\n')
```

**Insight:**
- Melakukan Cleaning dataset dengan melakukan pengisian NULL berdasarkan type datanya
- atribut bertipe Numerik menggunakan method fillna dan mean dimana akan mengisi dengan nilai mean (rata-rata)
- atribut bertipe Object menggunakan method fillna dan mode dimana akan mengisi dengan nilai mode (modulus)

**Melakukan perhitungan dan menambahkan atribut AQI (Air Quality Index) pada dataset yang ada**

```python
def calculate_aqi_pm25(pm25_value):
    if pm25_value <= 30:
        return pm25_value * 50 / 30
    elif pm25_value <= 60:
        return 50 + (pm25_value - 30) * 50 / 30
    elif pm25_value <= 90:
        return 100 + (pm25_value - 60) * 100 / 30
    elif pm25_value <= 120:
        return 200 + (pm25_value - 90) * 100 / 30
    elif pm25_value <= 250:
        return 300 + (pm25_value - 120) * 100 / 130
    elif pm25_value > 250:
        return 400 + (pm25_value - 250) * 100 / 130
    else:
        return 0

# Fungsi untuk menghitung AQI untuk PM10
def calculate_aqi_pm10(pm10_value):
    if pm10_value <= 50:
        return pm10_value
    elif pm10_value <= 100:
        return pm10_value
    elif pm10_value <= 250:
        return 100 + (pm10_value - 100) * 100 / 150
    elif pm10_value <= 350:
        return 200 + (pm10_value - 250)
    elif pm10_value <= 430:
        return 300 + (pm10_value - 350) * 100 / 80
    elif pm10_value > 430:
        return 400 + (pm10_value - 430) * 100 / 80
    else:
        return 0

# Fungsi untuk menghitung AQI untuk SO2
def calculate_aqi_so2(so2_value):
    if so2_value <= 40:
        return so2_value * 50 / 40
    elif so2_value <= 80:
        return 50 + (so2_value - 40) * 50 / 40
    elif so2_value <= 380:
        return 100 + (so2_value - 80) * 100 / 300
    elif so2_value <= 800:
        return 200 + (so2_value - 380) * 100 / 420
    elif so2_value <= 1600:
        return 300 + (so2_value - 800) * 100 / 800
    elif so2_value > 1600:
        return 400 + (so2_value - 1600) * 100 / 800
    else:
        return 0

# Fungsi untuk menghitung AQI untuk NO2
def calculate_aqi_no2(no2_value):
    if no2_value <= 40:
        return no2_value * 50 / 40
    elif no2_value <= 80:
        return 50 + (no2_value - 40) * 50 / 40
    elif no2_value <= 180:
        return 100 + (no2_value - 80) * 100 / 100
    elif no2_value <= 280:
        return 200 + (no2_value - 180) * 100 / 100
    elif no2_value <= 400:
        return 300 + (no2_value - 280) * 100 / 120
    elif no2_value > 400:
        return 400 + (no2_value - 400) * 100 / 120
    else:
        return 0

# Fungsi untuk menghitung AQI untuk CO
def calculate_aqi_co(co_value):
    if co_value <= 1:
        return co_value * 50 / 1
    elif co_value <= 2:
        return 50 + (co_value - 1) * 50 / 1
    elif co_value <= 10:
        return 100 + (co_value - 2) * 100 / 8
    elif co_value <= 17:
        return 200 + (co_value - 10) * 100 / 7
    elif co_value <= 34:
        return 300 + (co_value - 17) * 100 / 17
    elif co_value > 34:
        return 400 + (co_value - 34) * 100 / 17
    else:
        return 0

# Fungsi untuk menghitung AQI untuk O3
def calculate_aqi_o3(o3_value):
    if o3_value <= 50:
        return o3_value * 50 / 50
    elif o3_value <= 100:
        return 50 + (o3_value - 50) * 50 / 50
    elif o3_value <= 168:
        return 100 + (o3_value - 100) * 100 / 68
    elif o3_value <= 208:
        return 200 + (o3_value - 168) * 100 / 40
    elif o3_value <= 748:
        return 300 + (o3_value - 208) * 100 / 539
    elif o3_value > 748:
        return 400 + (o3_value - 400) * 100 / 539
    else:
        return 0

# Fungsi untuk menghitung AQI total berdasarkan semua polutan
def calculate_aqi(row):
    aqi_pm25 = calculate_aqi_pm25(row['PM2.5']) if pd.notna(row['PM2.5']) else np.nan
    aqi_pm10 = calculate_aqi_pm10(row['PM10']) if pd.notna(row['PM10']) else np.nan
    aqi_so2 = calculate_aqi_so2(row['SO2']) if pd.notna(row['SO2']) else np.nan
    aqi_no2 = calculate_aqi_no2(row['NO2']) if pd.notna(row['NO2']) else np.nan
    aqi_co = calculate_aqi_co(row['CO']) if pd.notna(row['CO']) else np.nan
    aqi_o3 = calculate_aqi_o3(row['O3']) if pd.notna(row['O3']) else np.nan
    return pd.Series({
        'AQI_PM25': aqi_pm25,
        'AQI_PM10': aqi_pm10,
        'AQI_SO2': aqi_so2,
        'AQI_NO2': aqi_no2,
        'AQI_CO': aqi_co,
        'AQI_O3': aqi_o3,
    })
df[['AQI_PM25', 'AQI_PM10', 'AQI_SO2', 'AQI_NO2', "AQI_CO", 'AQI_O3']] = df.apply(calculate_aqi, axis=1)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
mean_data = df.groupby("station")[["AQI"]].mean()
x = mean_data.index
y = mean_data["AQI"]

norm_y = (y - y.min()) / (y.max() - y.min())
colors = plt.cm.RdYlGn(1 - norm_y)
plt.bar(x, y, color=colors)
plt.xticks(rotation = 90)
plt.title('Rata-rata Kualitas udara di Stasiun')
plt.xlabel("Stasiun")
plt.ylabel("Mean kualitas udara")
plt.show()
```

**Didapatkan dari bar data diatas kualitas udara di stasiun Dingling memiliki kadar polutan terendah di bandingkan dengan stasiun yang lain**

```python
AQI_by_month = df.groupby('station').resample('M').mean().reset_index()

stations = AQI_by_month['station'].unique()

for station in stations:
    station_data = AQI_by_month[AQI_by_month['station'] == station]

    plt.figure(figsize=(10, 6))
    plt.plot(station_data['date'], station_data['AQI'], label=station)
    plt.title(f"Monthly AQI for Station: {station}")
    plt.xlabel('Date')
    plt.ylabel('AQI')
    plt.xticks(rotation=60, horizontalalignment='right', fontsize=10)

    plt.show()
```

**Insight:**
- Berdasarkan visualiasi data diatas didapatkan bahwa station yang memiliki kualitas udara yang cukup baik adalah stasiun Dingling dan disusul oleh Huairou
